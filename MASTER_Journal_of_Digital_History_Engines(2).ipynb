### 2.1 Theory: Functioning of Topic Modeling Algorithms

Topic Modeling is a probabilistic machine learning method that can be used to determine and extract topic areas in large document corpora. By calculating probability distributions, hypotheses are automatically generated and iteratively refined and verified. The method aims at identifying which words and word groups appear frequently in connection with other words and word groups. The algorithm does not use semantic knowledge, but calculates, independent of their meaning, which words are related to each other across a corpus and bundles them into groups - the topics. Conversely, this means that such topics can contain words that have nothing in common semantically. Based on the topics, it is possible in a next step to analyze which parts of the corpus - which documents - have a share in certain topics. Central in this context is the idea of a bag filled with (single) words, the bag of words. This also explicitly ignores the order of the words as well as linguistic or semantic dependencies. The only thing considered important is that the words occur in the same document. Topic Modeling has been used for a long time in literary studies and also in historical studies, but only in individual cases it is reflected which procedures lead to the extracted topics.

In these pages, we restrict ourselves to the quasi-standard of Latent Dirichlet Allocation (LDA). Although other topic modeling approaches exist, they are used less frequently due to the diverse research and existing literature on LDA. 

Currently, two programs are primarily used for topic modeling in digital humanities and computer science: Gensim, a Python library, on the one hand, and Mallet, a JAVA-based implementation, on the other. While Gensim is widely used in computer science, Mallet has a broader standing in the humanities, as there are excellent tutorials (e.g. programminghistorian.org) as well as implementations with graphical user interfaces (e.g. DARIAH Topics Explorer). In both approaches, direct intervention in the machine learning process is possible, whereas Gensim allows deep manipulation especially in (hyper)parameter tuning, which is important for machine learning applications. An advantage for the comparison of both engines is that there is a mallet wrapper for Gensim, which allows execution and (hyper-)parameter tuning of the JAVA engine within a Python environment.

### 2.3 Preprocessing: Preparing the Text


For all three corpora, regardless of the time period they cover or how extensive they are, processing beyond the digitization of the sources is necessary. These processing steps are commonly referred to as preprocessing. Since these steps represent a manipulation of the historical sources, this "data cleaning" has recently been the subject of intense critical discussion in the digital humanities. Accordingly, the steps performed here should not be understood as a standardized sequence, but should be applied in a reflective manner. Usually, lower casing (lower case of all characters), removal of punctuation marks and stopwords as well as, under certain circumstances, lemmatization (reduction of word forms to their basic form) are part of these steps. In the case of extremely long documents that cannot be further segmented, it is also useful to break them down into smaller parts, since Topic Modeling cannot optimally process long texts and documents of unequal length distort the analysis. The step necessary for this is called chunking.

It is worth going into more detail about so-called stopwords. Topic modeling algorithms, like various other text analysis methods, work with so-called stopword lists, which remove strings with potentially high occurrences and little or no semantic content. From an epistemological-hermeneutic perspective, the question must accordingly be asked where we identify meaning in textual elements and what criteria we use to do so. Against this background, the epistemological interest is also important, since it influences the assessment of whether a string is a priori definable as a stopword or whether a posteriori meaning loading becomes possible on the basis of its co-occurrence with other words within a topic. For each of the three corpora, different stopword lists were used or new ones were created.

<b>Chunking: Cut up texts</b>
One point is the repackaging of data into so-called dataframes, which can be processed more efficiently. Individual files, such as a pure text file for a chronicle, can be turned into packages that consist of a file name (first column) and the associated text (second column) in the same way as two-column tables. Dataframes have the further advantage that, for example, splitting into pieces can be easily accomplished. This so-called chunking helps that very long documents, for example the extensive interviews, can be divided into parts (chunks), which themselves can be understood as a unit of a topic model.

<b>Stopwords: Remove semantically meaningless words</b> 
Since topic modeling measures not only the ratio of word occurrences, but also the frequency of words, very frequently occurring words will also appear frequently in topics. Typically, this makes the interpretation of topics and the understanding of topics as topic fields difficult or even impossible. A remedy is the removal of words that are not or at most conditionally considered to be "meaning-bearing" via so-called stopword lists. The lists consist of single entries (nothing else than character strings), which are removed in preprocessing.

For modern language forms specific stop word lists exist, which are used. However, in corpora to be treated, words will always be found that do not function on such lists because they are dialectal - found, for example, in the interviews - or because they occur very frequently due to the formulaic nature of the documents and are considered meaningless, as in the case of the government council minutes. For pre-modern documents, the starting position is even more problematic, since there is no question of a standardization of the language.

The stop word lists trigger in themselves simple procedures, which can also be easily controlled. Nevertheless, the application is highly problematic, since the lists, as described above, have to be applied individually to the corpora and ultimately interfere massively with the content of the sources through the filtering process. In the sense of transparent and reproducible research, stop word lists belong to any application to topic modeling (co-)published. Only in this way it becomes reasonably comprehensible how the results are achieved and to what extent the composition of the sources has been interfered with.

<b>Lemmatization: operating with basic forms of words</b> 
Stop words are used to eliminate the potential influence of high frequency words. However, this is only one of the challenges in dealing with topic modeling. The engines/frameworks used for the processes are all not equipped with language knowledge and count the occurrence of words in terms of strings. This means that different word forms (such as genitive or plural) are understood as different words. This effect leads to the fact that relatively frequent words occur in different inflectional forms in one (or more) topics. 

To minimize linguistic homogeneity, texts can be lemmatized, i.e. the basic form of a word can be generated. This means that the texts (already without stop words) are modified in such a way that instead of the words the lemmas form the corpus to be processed. This procedure is carried out by matching the respective inflected word with a reference list, which contains all inflected forms of the complete vocabulary, if possible. If the word is found, it is replaced by the base form in the original text, thus counting, for example, "am", "are" and "was" as three times "to be". The reference list - the language model - in the pipeline used for this paper is Spacy, which has been trained on newspaper corpora and contains about 500,000 vectorized words. A painful desideratum from the perspective of historical scholarship is the lack of coverage of historical spellings in such language models. For example, virtually none are correctly recognized and resolved in medieval chronicles, few in 19th century council resolutions, and far from all words in dialectal interviews.

In order to make the effects of the described preprocessing comprehensible, in the following it will not only be demonstrated what results at best can be achieved with topic modeling, but also shown to what extent the processes lead to more "readable" and above all interpretable results.

### 3.2 Life story interviews

A qualitative comparison - which will now be demonstrated on the basis of the life history interviews - can also begin with a distant reading by first comparing topic lists with each other. During the evaluation of preprocessing and parameter tuning it turned out that a division of the extensive interviews (often several hours long) into shorter units (chunks) of 10 (Gensim) and 25 sentences (Mallet) results in consistent and meaningful topics. To determine the optimal number of topics, a simple metric was developed, which will be presented in detail in another paper, but is already included and commented in the published notebook: The "Keyword Diversity" (KD) is a simpler and easily comprehensible variant of the widely used Exclusivity or Relevance metrics, both closely related or derived from the Lift method. The KD calculates in percent how many keywords from all topics occur uniquely among the first n topwords of the topics and should thus be a first indicator for the discriminatory power of the topics. It has been shown that the keyword diversity settles at about 50 to 60 topics on a pedestal, therefore the models with 50 topics are compared here. 

How do the results of Gensim and Mallet relate to each other? A comparative review of the lists showed that Mallet delivers more convincing results overall. A comparative look at the highest weighted keyword of a topic alone shows that Gensim - which always had a significantly lower KD in the evaluation - delivers extremely redundant results: In five topics the word "war" is at the top, in eight others it is in second or third place. In total, "war" is among the ten most important keywords in 29 of 50 topics - in Mallet's case, however, only in six, three of them in first place. Another indicator of low selectivity is the ambiguous word "Essen", which stands for the Ruhr metropolis on the one hand and for food on the other. In the Gensim results, it is in first place in six topics, in second or third place in seven of the keywords, and in total it is found in the first ten keywords in 22 topics. In Mallet, it is found only once at the top of a topic and four times in second or third place; in total, it is represented only six times among the top ten keywords of the 50 topics. # 4. Engines applied [RECENT VERSION]

In all three corpora it became apparent that the choice of engine generates different results. The two methods are based on the same idea, the clustering of words that occur in similar contexts. However, this clustering is done differently and also requires different amounts of computing resources. A look at the methods shows that even small differences in the clustering process lead to strongly diverging results.

The main advantage of Topic Modeling is that the similarity of two randomly selected words in a document is calculated by starting at a random point and determining the probability of their occurrence. In subsequent steps, an attempt is made to identify "better", in the sense of more frequent, and correspondingly "more meaningful" combinations of word pairs. Due to the always slightly different starting point, the randomly chosen word pairs, the result will only be identical if the starting point (so called random seed or state) is fixed. From the methodological point of view, this is not unproblematic, since an own retracing is only possible if the starting point is firmly defined and openly communicated, and also if the identical corpus can be reused.

The main difference between the two engines Mallet and Gensim is the approximation of the a posteriori probability in Bayesian probability theory, which is calculated by so-called inference algorithms. After the random starting point is set as the first prior (typically a word and its relation to all other words), many small steps are taken to try to compute from it the a posteriori probability of occurrence with other words. Thus, to improve clustering, another word is taken to help and it is measured how likely it is that these two words could be part of a cluster. To elicit these and all other mini-clusters, there are different inference methods, at the core the two areas of sampling and optimization functions. 
While Mallet Gibbs uses sampling, Gensim relies on the more resource-efficient and faster, but effectively less precise Variational Bayes method. This assumes that all features (in Topic Modeling, "words") are independent and attempts to resolve the optimization function in continuous steps toward minimizing the Kullback-Leibler divergence between a priori and a posteriori distributions. 
Gibbs sampling, on the other hand, which is in the tradition of Markov Chain Monte Carlo (MCMC) methods, involves random sampling. In a kind of "walk through the vector space", randomly drawn samples, i.e. text segments, are gradually compared with each other and thus the a posteriori probabilities are calculated. Similar to n-grams, the last calculation is always used for comparison with the next one. In this way, a so-called Markov chain is generated, which is correspondingly more computationally expensive than the Variational Bayes method.
The big takeaway from Latent Dirichlet Allocation (LDA), which has become a quasi-standard among many topic modeling algorithms and forms the umbrella of these methods, is that each word and text segment can be assigned to more than just one cluster. The revelation of "latent" meaning structures of a text, according to Blei et al. can only be had if each entity can belong to multiple clusters. The serious differences between Gensim and Mallet ultimately stem from the fact that this layering of meaning levels relentlessly reveals the poor consistency and discriminatory power of topics.

```python
def ngrams_topics(data, n):


    words = data
  
    if n == '2':
        a = 0
        b = 1
        bigrams = []
        counter = 0

        for doc in words:
          for word in doc:
 
            if counter < len(words) - 1:
                if words[a] in stoplist or words[b] in stoplist:
                    a = a + 1
                    b = b + 1
                    counter = counter + 1
                else:
                    x = words[a] + ' ' + words[b]
                    bigrams.append(x)
                    a = a + 1
                    b = b + 1
                    counter = counter + 1
        
        return bigrams


    if n == '3':
        a = 0
        b = 1
        c = 2
        trigrams = []
        counter = 0

        for doc in words:
          for word in doc:
            while counter < len(words) - 2:
                if words[a] and words[b] and words[c] in stoplist:
                    a = a + 1
                    b = b + 1
                    c = c + 1
                    counter = counter + 1
                else:
                    x = words[a] + ' ' + words[b] + ' ' + words[c]
                    trigrams.append(x)
                    a = a + 1
                    b = b + 1
                    c = c + 1
                    counter = counter + 1


        return trigrams

#@markdown #Auswählen: Bi- oder Trigramme
n = "2" #@param [2,3]

#@markdown # Dataset auswählen
#@markdown data = voller Datensatz, normalisiert

#@markdown data_words_nostops = stopwortbereinigt

#@markdown data_final = stopwortbereinigt und lemmatisiert

dataset_ngrams = "data_words_nostops" #@param ['data', 'data_words_nostops', 'data_final']

if dataset_ngrams == 'data':
  dataset_ngrams = data
if dataset_ngrams == 'data_words_nostops':
  dataset_ngrams = data_words_nostops
if dataset_ngrams == 'data_final':
  dataset_ngrams = data_final

data_ngrams = [ngrams_topics(line, n) for line in dataset_ngrams] # ohne Lemmatisierung: data, line[1]

freq = {}

for doc in data_ngrams:
  for ngram in doc:
    if ngram not in freq:
      freq[ngram] = 1
    else:
      freq[ngram] += 1

freq_sorted = []

for ngram, count in freq.items():
    t = (count, ngram)
    freq_sorted.append(t)

freq_final_1 = sorted(freq_sorted, reverse=True)

pprint(freq_final_1[:249])

```

##Topword Diversity
Hier soll die Einzigartigkeit von Keywords innerhalb der Topics untersucht werden, in je weniger Topics ein Wort enthalten ist, desto trennschärfer sollten die Topics sein. Vgl. mit der Relevance Metric bei pyLDAvis, nähere Infos dazu im oben verlinkten Paper, und dem Exclusivity Score im Paper von Thompson/Mimno. Bislang harter Filter, es werden NUR einzigartige Wörter gezählt, ich möchte einen Threshold einbauen, dass "seltene" Wörter besser bewertet werden, also beispielsweise auch Wörter, die in 2-3 Topics vorkommen einen guten Score bekommen, Wörter, die in allen Topics vorkommen, einen schlechten. Das könnte man dann auch noch mit der jeweiligen Gewichtung im Topic verrechnen. Für Dennis hilfreich, weil er möglichst trennscharfe und konsistente Topics braucht, um die pluralistischen Inhalte der Sammlung abzubilden. -> evtl. noch in Zusammenhang mit Dictionary bringen
